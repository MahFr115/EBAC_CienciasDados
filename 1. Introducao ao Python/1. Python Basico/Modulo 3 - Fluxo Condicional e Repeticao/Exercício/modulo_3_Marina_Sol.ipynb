## 3\. Estrutura condicional try / except

3.1. Utilize a estrutura try/except para imprimir as cidades dos usuários que passaram mais de 70 segundos no site.

Dica: Realize o tratamento de exceções dentro da execução da iteração dos elementos da lista

```python
cid_try =[]
cid_type =[]
for dado_de_usuario in propaganda_online:

    try:
      if(dado_de_usuario['tempo_gasto_site'] > 70):
        cid_try.append(dado_de_usuario['cidade'])


    except TypeError:
      cid_type.append(dado_de_usuario['cidade'])

    except Exception:
      print("Erro na base de dados")

print("O site foi acessado por mais de 70 segundos em:\n")
for cidade in cid_try:
    print(cidade)

print("\n")

print("Não há dados relativos ao acesso ao site para:\n")
for cidade in cid_type:
    print(cidade)
```

1.2. Crie uma lista chamada `leads` com a renda dos usuários que clicaram na propaganda.

Dica: Para armazenar a renda dos usuários na lista `leads` utilize o método append().

```python
leads = []
key = []

for dado_de_usuario in propaganda_online:
  try:
    if (dado_de_usuario['clicou_no_ad'] == 1):
      leads.append(dado_de_usuario['renda_area'])

  except KeyError:
    key.append(dado_de_usuario['renda_area'])

print("O(s) usuário(s)", leads, "clicou(clicaram) na propaganda.\n")
print("Há um erro de chave para o(s) usuário(s)", key, ".\n")
```
